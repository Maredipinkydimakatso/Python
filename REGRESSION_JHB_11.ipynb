# Data engineering

```python
train_df.columns=[col.replace(' ','_') for col in train_df.columns]
train_df.head()
```

```python
#Dropping the 'Precipitation_in_millimeters' column since it has very few valid entries
# 199 non-null entries out of 7068
train_df=train_df.drop(['Precipitation_in_millimeters'], axis=1)
train_df.head()
```
